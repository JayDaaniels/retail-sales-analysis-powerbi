```python
import pandas as pd
import numpy as np

np.random.seed(42)

# -------- CUSTOMERS --------
n_customers = 5000
customers = pd.DataFrame({
    "Customer_ID": [f"C{str(i).zfill(4)}" for i in range(n_customers)],
    "Customer_Name": np.random.choice(
        ["John Doe", "jane doe", "J. Smith", "Alice", "bob", None],
        n_customers
    ),
    "Region": np.random.choice(
        ["North", "south", "EAST", "west", "West.", None],
        n_customers
    ),
    "Signup_Date": np.random.choice(
        pd.date_range("2020-01-01", "2023-12-31"),
        n_customers
    )
})

# -------- PRODUCTS --------
n_products = 500
products = pd.DataFrame({
    "Product_ID": [f"P{str(i).zfill(4)}" for i in range(n_products)],
    "Product_Name": np.random.choice(
        ["Laptop", "phone", "Chair", "Desk", "Tablet", None],
        n_products
    ),
    "Category": np.random.choice(
        ["Tech", "tech", "FURN", "Furniture", None],
        n_products
    ),
    "Cost_Price": np.random.uniform(50, 500, n_products),
    "Selling_Price": np.random.uniform(100, 1000, n_products)
})

# introduce bad pricing
products.loc[np.random.choice(products.index, 20), "Cost_Price"] = 900

# -------- ORDERS --------
n_orders = 60000
orders = pd.DataFrame({
    "Order_ID": np.random.choice([f"O{str(i).zfill(5)}" for i in range(n_orders//2)], n_orders),
    "Customer_ID": np.random.choice(customers["Customer_ID"].tolist() + [None], n_orders),
    "Product_ID": np.random.choice(products["Product_ID"].tolist(), n_orders),
    "Order_Date": np.random.choice(
        ["2023-01-01", "01/02/2023", "March 5 2023"],
        n_orders
    ),
    "Quantity": np.random.randint(1, 10, n_orders),
    "Discount": np.random.uniform(-0.2, 1.5, n_orders)
})

# -------- PAYMENTS --------
payments = pd.DataFrame({
    "Payment_ID": [f"PAY{str(i).zfill(5)}" for i in range(n_orders)],
    "Order_ID": np.random.choice(orders["Order_ID"], n_orders),
    "Payment_Method": np.random.choice(["Card", "Cash", "Transfer"], n_orders),
    "Payment_Status": np.random.choice(["Paid", "Pending", "Failed"], n_orders)
})

# -------- SAVE FILES --------
customers.to_csv("customers_brutal.csv", index=False)
products.to_csv("products_brutal.csv", index=False)
orders.to_csv("orders_brutal.csv", index=False)
payments.to_csv("payments_brutal.csv", index=False)

print("Brutal dataset created!")
```
